# Exploratory Data Analysis of Synthetic Data

## Plot temperature profiles

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/synthetic/house_1.csv')
plt.plot(df['timestamp'], df['outdoor_temp'])
plt.title("Outdoor Temperature - House 1")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Plot heating load profiles

plt.plot(df['fixed_load'])
plt.title("Fixed Load Profile - House 1")
plt.ylabel("kW")
plt.xlabel("Time step")
plt.show()

## Inspect comfort preferences

```python
print(df[['setpoint_temp', 'min_temp']].head(1))
```

## Compare multiple houses

```python
for i in range(1, 6):
    df = pd.read_csv(f'../data/synthetic/house_{i}.csv')
    plt.plot(df['fixed_load'], label=f'House {i}')
plt.legend()
plt.title("Fixed Loads of 5 Houses")
plt.show()
```
